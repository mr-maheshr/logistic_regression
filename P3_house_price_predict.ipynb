```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost import XGBRegressor
```

```python
# load data
hp_dataset = pd.read_csv('/content/Boston-house-price-data.csv')
```

```python
print(hp_dataset)
```

```python
# dataset -> dataframe
hp_dataframe = hp_dataset.copy()
```

```python
# print first 5 raws
hp_dataframe.head()
```

```python
# no of raws & columns
hp_dataframe.shape
```

```python
# check missing values
hp_dataframe.isnull().sum()
```

```python
# statistical measure
hp_dataframe.describe()
```

```python
# correlations in dataset
correlation = hp_dataframe.corr()
```

```python
# heatmap -> understand correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws= {'size':8}, cmap='Blues')
```

```python
# data split -> lable and feature
X = hp_dataframe.drop(['MEDV'], axis=1)
Y = hp_dataframe['MEDV']
```

```python
print(X)
print(Y)
```

```python
# data split -> train and test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# load model
xgb_model = XGBRegressor()
```

```python
# train model
xgb_model.fit(X_train, Y_train)
```

```python
# accuracy - train data prediction
train_data_prediction = xgb_model.predict(X_train)
print(train_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, train_data_prediction)

# Mean absolute error
score_2 = metrics.mean_absolute_error(Y_train, train_data_prediction)

print("R squared error (Train Data) : ", score_1) # closer to 0 better
print("Mean absolute error (Train Data) : ", score_2)
```

```python
# vsualize
plt.scatter(Y_train, train_data_prediction)
plt.xlable('Actual Price')
plt.ylable('Predicted Price')
plt.title('Actual Price vs Predicted Price')
plt.show()
```

```python
# accuracy - test data prediction
test_data_prediction = xgb_model.predict(X_test)
print(test_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean absolute error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error (Test Data) : ", score_1) # closer to 0 better
print("Mean absolute error (Test Data) : ", score_2)
```
